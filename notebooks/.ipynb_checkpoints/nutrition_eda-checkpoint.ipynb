```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

import os
from dotenv import load_dotenv
load_dotenv()


api_key = os.getenv('USDA_API_KEY')
url = "https://fdc.nal.usda.gov/api/v1/foods/search"

params = {
    'query': 'apple',
    'pageSize': 100,
    'api_key': api_key
}

r = requests.get(url, params=params)
foods = r.json()['foods']
print(foods)
```
